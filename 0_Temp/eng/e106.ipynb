```python
# variable length argument
def vfunc(*args):
  print(type(args))
  for arg in args:
    print(arg)
  
def vfund(**kwargs):
  print(type(kwargs))
  for k, v in kwargs.items():
    print(f"{k} = {v}")
    
v1 = 1; v2 = 2; v3 = 3
# k = [{"name":"James", "age": 99999}
# k1 = "James"; k2 = "9999"
vfunc(v1, v2, v3)
# vfund(k)
vfund(k1="James", k2=99999)
    
```

```python
class Math:
  @staticmethod
  def add(a, b):
    return a + b
  
  @staticmethod
  def multiply(a, b):
    return a * b
  
r = Math.add(1,2)
print(r)
```
