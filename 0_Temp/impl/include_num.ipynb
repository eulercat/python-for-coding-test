```python
n = int(input())
check_str = '3'
count = 0
for i in range(n+1):
    if check_str in str(i):
        count += 1     
print("count: %d" % count)

count_in_60 = 0
for i in range(60):
    if check_str in str(i):
        print("%d" % i)
        count_in_60 += 1
print("count_in_60: %d" % count_in_60)

# count = count * count_in_60 * count_in_60
a = count * 60 * 60
b = (n+1) * count_in_60 * 60
c = (n+1) * 60 * count_in_60

count = a + b + c

print(count)
```

```python
n = int(input())
check_str = '3'
count = 0
for i in range(n+1):
    for j in range(60):
        for k in range(60):
            if check_str in str(i) + str(j) + str(k):
                print("%d%d%d" % (i, j, k))
                count += 1
print("count: %d" % count)
print(count)
```
